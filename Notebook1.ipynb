```python
# Checking for missing values
meta_missing = df_meta.isna().mean().sort_values(ascending=False)
review_missing = df_review.isna().mean().sort_values(ascending=False)

print("Missing meta:\n", meta_missing.head(15))
print("Missing review:\n", review_missing.head(15))
```

***Understanding the missing value, other than the null value that are most present in price and in bought_together. We need to analyze even the value that don't have anything write***

```python
# Converting all the missing value that are written as [] or {} to NaN

def empty_to_nan(x):
    if x == [] or x == {}:
        return np.nan
    return x

# pandas 3.0 removed `DataFrame.applymap`, so we fall back to applying
# the transformation on each column. This achieves the same element-wise
# behavior without relying on the removed method.

df_meta = df_meta.apply(lambda col: col.apply(empty_to_nan))
df_review = df_review.apply(lambda col: col.apply(empty_to_nan))
```

```python
df_meta.replace("", np.nan, inplace=True)
df_review.replace("", np.nan, inplace=True)
```

```python
meta_missing = df_meta.isna().mean().sort_values(ascending=False)
review_missing = df_review.isna().mean().sort_values(ascending=False)

print("Meta missing ratio:\n", meta_missing.head(20))
print("Review missing ratio:\n", review_missing.head(20))
```
